```python
def format_us_number(num):
    num = str(num).replace('.0', '')         # Remove decimal point if present
    num = ''.join(filter(str.isdigit, num))  # Keep digits only
    if len(num) == 10:
        return f'+1 {num[:3]}-{num[3:6]}-{num[6:]}'
    else:
          return 'Invalid number'

contact_df['phone'] = contact_df['phone'].apply(format_us_number)
```
